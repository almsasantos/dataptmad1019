## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Use `Months_encoded` and `Week_or Weekend` columns
- Use `matplolib.pyplot` object oriented API
- Set your size figure to `12,4`
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures

```python
# Driver 1 enter codes here

#To get the median steps per month
median = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()
median

#List of months (1 to 12) to pass to the bar plot
months = fitbit['Months_encoded'].unique()
sorted(months)

#The values needed to build the second graphic:
analysis2 = {'workdays': fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()[1],'weekend_days': fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()[0]}
lists = sorted(analysis2.items())
x, y = zip(*lists)

#Create the figure for 2 graphs:
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

ax[0].plot(sorted(months), median, color='chocolate')
ax[0].set_title('Median of steps per month')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Median of steps')
#ax[0] = plt.xticks(list(range(13)),['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

ax[1].bar(x, y, color=['yellowgreen', 'chocolate'])
ax[1].set_title('Median of steps per week day')
ax[1].set_xlabel('Week day')
ax[1].set_ylabel('Median of steps')
```

```python
# Relevant code to resolve the second exercise of challenge 2:

data=['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
colors = ['darkred', 'darkblue', 'mediumvioletred']
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20,12))
for i in range(0, 3):
    ax[i].scatter(fitbit[data[i]], fitbit['Steps'], color=colors[i])
    ax[i].set_title(f'{data[i]} vs Steps')
    ax[i].set_xlabel(data[i])
    ax[i].set_ylabel('Steps')
```
