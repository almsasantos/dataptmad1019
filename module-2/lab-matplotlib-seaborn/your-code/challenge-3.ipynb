#### Whether the bin size affects your plot? Why?

```python
#your comment here
'''

The bin size affects our plot because less bins will receive more data per bin, as we can see with the distribution of age, 
when bins equals to 10 we'll have more data between age range of 25 to 34.9. When we change the bins value to 20, the range of
ages change to 27.5-32.49.

On the other hand, when bins value is 50 we'll have more accurate information, and the maximum of age distribution visualized
in this graphic belongs to range 22-23,999.

So once you're doing a statistics studies it's really important to always use the same amount of bins and to have the data
information as accurate as possible.

'''
```

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax = sns.heatmap(titanic.corr(method='pearson'), annot=True, fmt='0.2f')
ax.set_title('Correlation matrix')
```

#### Same resolution using cufflinks and plotly.express

```python
# Using cufflinks:
titanic.corr().iplot(kind='heatmap', colorscale='Blues', title='Correlation Matrix')
```

```python
# Using plotly.express:
import plotly.graph_objects as go
fig = go.Figure(data=go.Heatmap(z=titanic.corr()))
fig.show()
```
