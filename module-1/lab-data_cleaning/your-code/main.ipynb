#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
# Low variance columns:

low_variance = []


for col in merge_df._get_numeric_data():
    minimum = min(merge_df[col])
    ninety_perc = np.percentile(merge_df[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)
        
        
print(low_variance)
```

```python
# After cheking how many missing values exists in which columns, we should remove the columns
# filled with more than 40% of null values

threshold = 40
null_cols = null_cols[null_cols > 0]/len(merge_df)*100
null_filter = null_cols > threshold
drop_cols = list(null_cols[null_filter].index)


merge_df.drop(drop_cols, axis=1, inplace=True)


# Since there aren't negative Ids in the real word, we should remove all rows that have negative 
# values in certain columns as seen below.
no_negative_values = merge_df[(merge_df['userId'] <0) | (merge_df['AccountId'] <0) | (merge_df['postId']<0)].index
merge_df.drop(no_negative_values, inplace=True)
merge_df
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
# The first step to check the outliers is to calculate the interquartile that results from the subtraction 
# of the 75% percentile for the 25% percentile.

stats = merge_df.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=merge_df.columns)


for col in stats.index:
    iqr = stats.at[col, 'IQR']
    cutoff = iqr *1.5
    lower = stats.at[col, '25%'] - cutoff
    upper = stats.at[col, '75%'] + cutoff
    results = merge_df[(merge_df[col] < lower) | (merge_df[col] > upper)].copy()
    results['outliers'] = col
    outliers = outliers.append(results)   
    
    
# To save the outliers dataframe into a cvs folder use the following command:
outliers.to_csv


display(outliers) 
```
